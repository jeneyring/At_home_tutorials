```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Data

```python
print(sns.get_dataset_names()) #seaborns built-in data sets
```

### load data set

```python
crash_df = sns.load_dataset('car_crashes')
crash_df. head()
```

### Distribution Plot

```python
# a uni-vari distribution (one variable)
sns.distplot(crash_df['not_distracted'], kde=True, bins=25,)
```

```python
## Notes on additional variable above: 
##kde: with a kernal density estimation line (distribution of data)
## bins divide variables into bins
```

### Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
##notes on joint plot above:
#Creating a plot line based on if variable was a car crash that involved speeding an alcohol (for x & y)
#data is what data frame is being accessed/referred
#kind is what line is wanting (reg is a regression line)
```

### KDE Plots

```python
sns.kdeplot(crash_df['alcohol'])
```

### Pair Plots

```python
#plots relationship between all data of numerical values
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

### Rug Plot

```python
sns.rugplot(tips_df['tip']) #plots a single plot of data points as sticks (density of lines is where data is most common)
```

### Styling

```python
sns.set_style('darkgrid') #can use whitegrid, whites, darks, ticks, etc...

plt.figure(figsize=(8,4))

sns.set_context('paper', font_scale=1.6) #besides paper you can use talk and poster

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=True) #takes off spine on side you set
```

## Categorical Plots


### Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)  #the line references the variances of the data
```

```python
##notes:
#estimator lets you choose what data is based on. np.median is the median of the data
```

### Count Plot

```python
##counts the number of occurences

sns.countplot(x='sex', data=tips_df)
```

### Box Plot

```python
#compares different variables. Show quartiles of different data
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=0) ## allows you can relocate legends so it doesn't overstack on data
```

```python
## the bar across the box is the median 
## the box extends one stddev from the median
## the whiskers extend to the other data beyond the stddev
## diamonds are the outliners
```

### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
##violin plot uses the kde estimation 
```

### Strip Plot

```python
plt.figure(figsize=(8,5))
            
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
            
#shows different distributions with boxplots
#jitter helps spread the data points off of one another
#dodge gives each sex their own column
```

### Swarm Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
##swarm plots will stack on violin plots 
```

### Palettes

```python
plt.figure(figsize=(8,7))

sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='Wistia')
#look for color maps in matplotlib.org for palettes

plt.legend(loc=1)#to move the legend, upper_right is: 1, upper_left: 2, lower_left: 3, lower_right: 4
```

### Heat maps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx=crash_df.corr()#corr will tell you how infuential the data is on the result (negatives mean no real correlation)
crash_mx
```

```python
sns.heatmap(crash_mx, annot=True, cmap="Blues")
```

```python
##or, another way to create this...

flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month',columns='year', values='passengers')
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
```

### Cluster Maps

```python
iris= sns.load_dataset('iris')
species=iris.pop('species') #we want to return values based on species
sns.clustermap(iris)
#this is a heirarchally clustered heat map. it compares columns and rows of heat map and clusters the close data points
```

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1) 
#standard_scale normalizes the data to focus on the clustering (messes up order of data (see years))
#the heat map does not reposition data while cluster map tightly fits data so you can see patterns
```

### Pair Grid

```python
#like pair plots, where different plots are shown, while pair grid lets you get more specific
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map_diag(plt.hist)
#iris_g.map_offdiag(plt.scatter) #can be specific on which plot shows what
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
iris_g=sns.PairGrid(iris, hue='species',
                    x_vars=['sepal_length', 'sepal_width'],
                   y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

### Facet Grid 
### print multiple plots on a grid where you can define columns and rows

```python
tips_fg=sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_fg=sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                     col_order=['Dinner','Lunch'], palette='Set1') #aspect width based on height
tips_fg.map(plt.scatter, 'total_bill', 'tip',edgecolor='k')
```

```python
tips_fg=sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                     col_order=['Dinner','Lunch'], palette='Set1') #aspect width based on height
tips_fg.map(plt.scatter, 'total_bill', 'tip',edgecolor='k')

```

```python
kws=dict(s=50, linewidth=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex',hue='smoker',height=4,aspect=1.3,
                       hue_order=["Yes","No"],
                       hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention') 
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)#each person is in their own plot and compared
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```
